Python Datatypes 

```python
#1. integer - It is a whole number, it can be either positive, negative or zero but it does not include any fractional part. e.g 10,14,1234566, -43, -587.
num1 = 100
print(num1)
print(type(num1))
```

```python
# 2. float - it is a decimal number. e.g 23.0 , 44.4324 , -123.2341.

num2 = 100.00
print(num2)
print(type(num2))
```

```python
#  3. complex -  It is used for working with complex numbers, here the imaginary part is represented by 'j'.

complex_number = 45 + 6j
print(complex_number)
type(complex_number)
```

```python
#  4. string - It is a collection of characters (numbers, alphabets, special symbols and spaces) kept in single or double quotes. It can have any number of characters or it can be empty also. 

string1 = "This is a string"
string2 = 'A'
string3 = "word"
string4 = ""
string5 = ''' This is a 
            multiline
            string. '''
print(type(string5))
```

```python
# 5. list : It is a heterogenous collection of data items seperated by comma. It is mutable and represented by []

list1 = [10,12.33,14+3j, "string",(1,2,3,4),["a","b","c"]]
print(list1)
print(type(list1))
```

```python
# 6. tuple : It is a heterogenous collection of data items seperated by comma. It is immutable and represented by ()

tuple1 = (10,20,30,40,50,[1,2,3,4])
print(type(tuple1))
```

```python
# 7. Boolean : It is a datatype with two values i.e True and False. 

result = True
print(type(result))
```

```python
# 8. Set : It is an unordered collection of unique elements/data items, It is mutable and does not allow any duplicates.
set1 = {10,13,14,15.00,20,10,10,10,20,20,30,432,453,564,13}
print(set1)
print(type(set1))
```

```python
# 9. Dictionary (dict) : It is an collection of data items which are kept in pairs (key-value pairs) 

dict1 = {
    1 : "One",
    2 : 20,
    3 : 40.00,
    "four" : [10,20,30,40],
    "five" : 5
}

print(dict1)
type(dict1)
```

```python
print(dict1["five"])
```

```python
# Binary Datatypes 
number1 = 12345
bytes1 = bytes(number1)
print(bytes1)
print(type(bytes1))
```

```python
# memoryview
m = memoryview(bytes1)
print(m)
```

```python
# bytearray 
array = bytearray(number1)
print(array)
type(array)
```
